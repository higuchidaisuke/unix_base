```python
def readlines_file(file_name):
    # 行毎のリストを返す
    with open(file_name, 'r') as file:
        return file.readlines()


def  save_file(file_name, text):
    with open(file_name, 'w') as file:
        return file.write(text)
cal10 = readlines_file('col5.txt')
cal11 = readlines_file('col6.txt')

# 改行や空白文字を削除
cal10 = list(map(lambda x: x.strip(), cal10))
cal11= list(map(lambda x: x.strip(), cal11))

# タブ区切りで並べたリストを作成
lines = ["{0}\t{1}\n".format(line1, line2) for line1, line2 in zip(cal10, cal11)]
save_file('margo.txt', "".join(lines))
with open('margo.txt') as f:
    print(f.read())
```

```python
import math

fname = 'hightemp.txt'
n = int(input('N--> '))

with open(fname) as data_file:
    lines = data_file.readlines()

count = len(lines)
unit = math.ceil(count / n)  

for i, offset in enumerate(list(range(0, count, unit)), 1):
    with open('a{}'.format(i), mode='w') as out_file:
        for line in lines[offset:offset + unit]:
            out_file.write(line)
for i in range(1,len(list(range(0,count,unit)))+1):
    with open('a%d'%i) as f:
        print(f.read())
```
